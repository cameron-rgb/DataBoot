```python
#using graph
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('Video_Games_Sales.csv')

#sums up all values by Year of Release
df = df.groupby(['Year_of_Release']).sum()
#grabs only the global sales
df = df['Global_Sales']
#graphs
df.plot.line()
```

```python
#Question: What is the top and bottom platform by global sales?
import pandas as pd
df=pd.read_csv('Video_Games_Sales.csv')

#Sorts the summed total sales per platform by descending values
df = df.groupby(['Platform']).sum().sort_values('Global_Sales', ascending=False)
#Calls only global sales for a cleaner view
#df.head(6)['Global_Sales']
#And bottom sales 
df.tail(6)['Global_Sales']
```

```python
# Which developer has the highest average critic review score?
# import data
import pandas as pd
df=pd.read_csv('Video_Games_Sales.csv')

# get the average of each collum per Developer and sort the new table based on highest avg critic review score
df = df.groupby(['Developer']).mean().sort_values('Critic_Score', ascending=False)
#grab only relivant info and display top developers
df = df.head(3)['Critic_Score']
df.plot.bar(ylabel='Average Critic Review Score')
```
